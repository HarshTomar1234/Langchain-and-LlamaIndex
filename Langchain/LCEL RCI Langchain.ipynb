RCI (Recursive Criticism and Improvement) is a method in LangChain that allows for the creation of chains that recursively criticize and improve their own output. This approach is particularly useful for automating computer tasks and enhancing the performance of large language models (LLMs).

### Key Components

1. **Criticism**: The RCI chain critiques its own output, identifying problems and areas for improvement.
2. **Improvement**: Based on the identified problems, the chain generates an updated output that addresses these issues.

### Implementation

RCI chains can be implemented using the LangChain Expression Language (LCEL). This involves defining the chain structure, including the prompts, models, and output parsers. The chain is then executed recursively, with each iteration refining the output based on the previous iteration's criticism.

### Example

Here is an example of how an RCI chain might be implemented using LCEL:
```python
from langchain import LCEL

# Define the RCI chain
chain = LCEL(
    prompt="Write a short story about a cat.",
    model="llama-2",
    output_parser="str"
)

# Run the chain recursively
output = chain.invoke({"input": "Write a short story about a cat."})

# Critique the output
critique = chain.invoke({"input": "Critique the previous output."})

# Improve the output based on the critique
improved_output = chain.invoke({"input": "Improve the previous output based on the critique."})

print(improved_output)
```
### Benefits

RCI chains offer several benefits, including:

- **Improved Accuracy**: By recursively criticizing and improving its own output, the chain can refine its results to achieve higher accuracy.
- **Flexibility**: RCI chains can be applied to a wide range of tasks and can be easily integrated with other LangChain components.
- **Efficiency**: The recursive approach allows the chain to refine its output without requiring extensive manual intervention.

### Integration with LangChain

RCI chains are fully integrated with the LangChain framework, allowing for seamless interaction with other LangChain components and tools. This integration enables developers to leverage the full power of LangChain to create sophisticated AI applications.

### Common Issues

- **Infinite Loops**: One common issue with RCI chains is the risk of infinite loops, where the chain continues to critique and improve its output indefinitely. This can be mitigated by implementing stopping criteria or limits on the number of iterations.
- **Model Overfitting**: Another issue is the risk of model overfitting, where the chain becomes overly specialized to the specific task and loses generalizability. This can be addressed by using techniques such as regularization or early stopping.

````python
prompt = ChatPromptTemplate.from_template(
    "tell me an intersting fact about {subject}"
    )

reverse_prompt = ChatPromptTemplate.from_template(
    "based on this interesting fact which is chunked down from a meta subject:\n\n {interesting_fact}\n\n Recover what the meta subject is\n Subject:"
    )
````

# As LCEL

````python
langchain.debug = False
````

````python
from langchain import PromptTemplate
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    AIMessagePromptTemplate,
    HumanMessagePromptTemplate,
)
````

````python
template="You are a helpful assistant that imparts wisdom and guides people with accurate answers."
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template="{question}"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
````

````python
chain1 = chat_prompt | model | StrOutputParser()
````

````python
initial_question = "Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"
````

````python
initial_answer = chain1.invoke({"question": initial_question})
initial_answer
````

````python
fake_initial_ai_answer = """Roger initially has 5 tennis balls. Each can of tennis balls contains 3 tennis balls, and he bought 2 cans, so he has 2 x 3 = 6 additional tennis balls.
Therefore, the total number of tennis balls Roger has now is 5 + 4 = 9."""
````

## Part 2 - Critique  

````python
template="You are a helpful assistant that looks at answers and finds what is wrong with them based on the original question given."
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template="### Question:\n\n{question}\n\n ###Answer Given:{initial_answer}\n\n Review your previous answer and find problems with your answer"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
````

````python
rc_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
````

````python
chain2 = rc_prompt | model | StrOutputParser()
````

````python
constructive_criticism = chain2.invoke({"question": initial_question, "initial_answer":fake_initial_ai_answer})
constructive_criticism
````

## Part 3 - The Improvement

````python
template="You are a helpful assistant that reviews answers and critiques based on the original question given and write a new improved final answer."
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template="### Question:\n\n{question}\n\n ###Answer Given:{initial_answer}\n\n \
###Constructive Criticism:{constructive_criticism}\n\n Based on the problems you found, improve your answer.\n\n### Final Answer:"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
````

````python
improvement_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
````

````python
chain3 = improvement_prompt | model | StrOutputParser()
````

````python
final_result = chain3.invoke({"question": initial_question,
                              "initial_answer":fake_initial_ai_answer,
                              "constructive_criticism": constructive_criticism})

final_result
````
